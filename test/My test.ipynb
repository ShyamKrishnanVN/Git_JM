```
import time

# Calculate the end time (current time + 5 minutes)
end_time = time.time() + 300  # 300 seconds = 5 minutes

# Run the script for 5 minutes
while time.time() < end_time:
    print("This script is running for 5 minutes.")
    time.sleep(1)  # Sleep for 1 second between each iteration
```

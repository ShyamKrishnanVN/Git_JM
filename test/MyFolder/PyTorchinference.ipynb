```
import torch
import torch.nn as nn
from sklearn.preprocessing import StandardScaler

# Load the saved model
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(4, 10)
        self.fc2 = nn.Linear(10, 3)  # 3 output classes for Iris dataset

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = self.fc2(x)
        return x

model = Net()
model.load_state_dict(torch.load('/nfsdir/uploadfiles/1156/1130855202/model/1925219678/Pytorch_model_Extrenal_V1.pth'))
model.eval()

# Define a function for making predictions
def predict_iris(features):
    scaler = StandardScaler()
    scaled_features = scaler.fit_transform(features);
    with torch.no_grad():
        inputs = torch.tensor(scaled_features, dtype=torch.float32)
        outputs = model(inputs)
        _, predicted = torch.max(outputs, 1)
    return predicted.numpy()

# Example usage:
# Features for prediction (replace with your own data)
features = [[5.1, 3.5, 1.4, 0.2],  # Example Iris Setosa
            [7.0, 3.2, 4.7, 1.4],  # Example Iris Versicolor
            [6.3, 3.3, 6.0, 2.5]]  # Example Iris Virginica

# Make predictions
predictions = predict_iris(features)

# Print the predictions
for idx, pred in enumerate(predictions):
    print(f'Example {idx + 1}: Predicted class = {pred}')
```
